### Non-Medical Dataset 2

```python
non_medical_dataset_2 = load_dataset("nbertagnolli/counsel-chat")
non_medical_dataset_2
```

```python
non_medical_dataset_2['train']['questionText'][2]
```

```python
questionID_list = []
unique_entries_non_medical_2 = []

for idx, questionID in enumerate(non_medical_dataset_2['train']['questionID']):
    if questionID not in questionID_list:
        questionID_list.append(questionID)
        unique_entries_non_medical_2.append(non_medical_dataset_2['train']['questionText'][idx])
```

```python
len(unique_entries_non_medical_2)
```

```python
unique_entries_non_medical_2[1]
```

```python
df_4 = pd.DataFrame({
    'questions' : unique_entries_non_medical_2,
    'label' : 'non-medical'
})
```

```python
df_4.head()
```

```python
df_4.shape
```

```python
final_dataset = pd.concat([final_dataset, df_4], ignore_index=True)
```

```python
final_dataset.shape
```

### Non-Medical Dataset 3

```python
non_medical_dataset_3  = load_dataset("nreimers/reddit_question_best_answers", split="train", streaming=True)
non_medical_dataset_3
```

```python
from itertools import islice

sample_data = list(islice(non_medical_dataset_3, 15000))
```

```python
len(sample_data)
```

```python
sample_data[0]['title']
```

```python
sample_data[0]['body']
```

```python
combined_text = sample_data[0]['title'] + ' ' + sample_data[0]['body']
cleaned_text = combined_text.replace('\n', ' ').replace('*', ' ')
```

```python
cleaned_text
```

```python
cleaned_texts = []

for data in sample_data:
    combined_text = data['title'] + ' ' + data['body']
    cleaned_text = combined_text.replace('\n', ' ').replace('*', ' ').strip()
    cleaned_texts.append(cleaned_text)
```

```python
len(cleaned_texts)
```

```python
cleaned_texts[2]
```

```python
df_5 = pd.DataFrame({
    'questions' : cleaned_texts,
    'label' : 'non-medical'
})
```

```python
df_5.head()
```

```python
df_5.shape
```

```python
final_dataset = pd.concat([final_dataset, df_5], ignore_index=True)
```

```python
final_dataset.shape
```
